## 인식 방법과 표현 방법

과거: 데이터의 크기와 다양성이 문제되지 않았음. -> 데이터를 이해하고 분석하는 것이 간단한 과정이었음.

현재: 엄청나게 많은 데이터가 셀수 없는 영역에서 존재.

시각화: 사람이 정보를 인지하는 데 도움을 주며, 시각화된 데이터와 인터랙팅하는데 도움을 줌.
* 과학의 시각화: 고유한 물리적 개체를 가지는 과학적 데이터양을 포함
* 정보의 시각화: 추상적인 정보를 포함

대부분의 시각화 시스템은 사람과 컴퓨터가 협력하고 아래 업무들을 수행할 수 있게 구현돼 있다.
* 데이터 분석 향상을 위해 시각적으로 표현
* 데이터 이해, 아이디어, 가설, 통찰을 위한 시각적 표현 모델
* 가설을 보조해주거나 모순되는 증거를 찾기 위해 사용자의 모델을 향상시켜주는 것
* 아이디어를 공유하고 조직화하기 위해 사용자를 도와주는 것

또한 데이터 시각화의 큰 장점 중 하나는 언어적 정보보다 시각적 정보를 빠르게 처리하는 능력을 향상시키는 것.
1920년 대 독일 심리학자들은 지각 조직화를 연구헀고, 그 첫번쨰 그룹은 게슈탈트 이론가들이었다.

## 지각에서의 게슈탈트 법칙

#### 게슈탈트
게슈탈트: '조직화된 전체' 혹은 개별적으로 식별된 부분들은 전체에 대한 서로 다른 특징을 가진다는 것을 의미.

* 근접성(Proximity): 물체가 서로 가까이 있거나 서로 연결돼 있을 떄, 그룹으로 인식해, 물체들의 분리를 줄이려고 함

<img src="그림0.png" width = 500/> 

* 유사성(Similarity): 물체들이 비슷한 속성, 색, 형태를 가질떄 같은 그룹으로 인식함

<img src="그림1.png" width = 500/> 
* 공동 운명(Common fate): 한 공간에 근접성과 유사성을 모두 만족하는 물체들이 있을 경우, 물체들의 방향성을 기준으로 다시 그룹화 함

<img src="그림2.png" width = 500/> 

* 좋은 연속성(Good continuation): 어떤 것들은 전체로서 중요함. 만약 끊어진 부분이 있다면, 직관적으로 이해.(2개의 교차된 선으로 인식)

<img src="그림3.png" width = 500/> 

* 폐쇄성(Closure): 부분의 경계가 손실되더라도 우리는 틈새를 무시하고 경계가 연결된 하나의 모양으로 봄

### 인터랙티브 시각화
시각화를 위해서는 상호작용이 반드시 고려되야 함
- 사람의 입력: 정보를 시각적으로 표현하는 일부 제어 권한은 사람이 조절할 수 있어야 함

- 반응 시간: 사람에 의해 변경된 부분은 적절한 시간 후에 반드시 시각화 정보에 반영되야 함

상호작용의 장점: 많은 양의 정보를 하나의 플랫폼을 통해 이해할 수 있게 해 짧은 시간안에 탐색할 수 있게 도와줌.

모든 시각화의 방법은 좋은 레이아웃 계획이 반드시 계획되어야 한다.

상호적인 시각화는 이벤트 리스너를 이용해 구현됨
